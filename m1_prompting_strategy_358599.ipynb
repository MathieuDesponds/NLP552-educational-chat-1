### Tactics

Before starting to prompt, I watched a short course called "ChatGPT Prompt Engineering for Developers" by Andrew Ng (https://www.deeplearning.ai/short-courses/chatgpt-prompt-engineering-for-developers/). In this course, the authors describe best practices for prompting. Here are a few key tactics:

#### Tactic 1: Use delimiters to indicate distinct parts of the input
 - Delimiters can be anything, such as ```, """, < >, <tag> </tag>, or :.
#### Tactic 2: Ask for structured output
 - Use custom or well-known structures (such as JSON or HTML).
#### Tactic 3: Ask think step-by-step:
 - E.g. think step-by-step, Go step by step through all possible answers
#### Tactic 4: Specify the steps required to complete a task
 - Write an accurate and clear step-by-step plan for the model to complete a task.


There are two types of questions in this task: Multiple Choice Questions (MCQ) and Open-ended Questions. For each of these types of questions, I will write different prompts. Let's start with the basic prompts and gradually improve them by adding the tactics described above.

````python
def usePrompt(prompt_OQ, prompt_MCQ):
    for i, test in enumerate(data[:3]) :
        chat = Chat.create("Question number {}".format(i))
        if 'choices' in test:
            prompt = prompt_MCQ.format(test['question'], test['choices'])
        else:
            prompt = prompt_OQ.format(test['question'])
        message = chat.ask(prompt).to_dict()
        print('\n\n\n-------------Question {}----------------\n\n\n'.format(i))
        print('PROMPT: ', prompt)
        print('\nASSISTANT REPLY\n')
        print(message['content'])
````

## Tactics 1 and 2.

````python
prompt_OQ = """
You are taking an exam for a university course. You will be given a question and your task is to answer correctly to the given question.
Use the following format:
Question:
```
question here
```
Correct Answer:
```
your answer here
```
Explanation:
```
here is an explanation and justification of your answer. 
```
Question:
```
{}
``` 
Correct Answer:
"""
````

````python
prompt_MCQ = """
You are taking an exam for a university course. You will be given a multiple-choice question and your task is to select only one correct answer from the offered list of options. 
Use the following format:
Question:
```
question here
```
Choices:
```
here is a list of possible answers
```
Correct Answer:
``` Here is the correct answer that you chose from the offered list. format:
variant number. answer ```

Explanation:
```
Here is an explanation and justification for each option why it's true or false. format:
variant number. "TRUE" or "FALSE": Your explanation
```

Question:
```
{}
``` 
Choices:
```
{}
```
"""
````

# 2: Tactics 3 and 4, to improve the quality of answers.

````python
prompt_OQ = """
You are taking an exam for a university course. You will be given a question and your task is to answer correctly to the given question. 

To answer the question, try to think step-by-step and write clear explanations for your answers using no more than 50 words.
 
Use the following format:
Question:
```
question here
```
Correct Answer:
```
your answer here
```
Explanation:
```
here is an explanation and justification of your answer. 
```
Question:
```
{}
``` 
Correct Answer:
"""
````

````python

prompt_MCQ = """
You are taking an exam for a university course. You will be given a multiple-choice question and your task is to select only one correct answer from the offered list of options. 

To solve the problem, follow these steps:
- Go step by step through all possible answers from the given list of options.
- For each option, try to think whether this answer is correct or not. Is this a complete and appropriate answer to the question asked?
- When you have checked all the answers from the given list of options, choose the most complete and appropriate answer among them. Do not choose the correct answer until you have checked all the answers from the given list.

Please, for each option from the offered list write an explanation of why you think it's false or true using no more than 50 words.
 
Use the following format:
Question:
```
question here
```
Choices:
```
here is a list of possible answers
```
Correct Answer:
``` Here is the correct answer that you chose from the offered list. format:
variant number. answer ```

Explanation:
```
Here is an explanation and justification for each option why it's true or false. format:
variant number. "TRUE" or "FALSE": Your explanation
```

Question:
```
{}
``` 
Choices:
```
{}
```
"""
````

````python
usePrompt(prompt_OQ, prompt_MCQ)
````

**Now the answers are structured, the model generates more better answers.**

````python
import json
import time
import numpy as np

all_iteractions = []

for i, test in enumerate(data) :
    time.sleep(3)
    chat = Chat.create("Question number {}".format(i))
    if 'choices' in test:
        prompt = prompt_MCQ.format(test['question'], test['choices'])
    else:
        prompt = prompt_OQ.format(test['question'])
        
    message = chat.ask(prompt).to_dict()
    interaction = [{'role': 'user', "content": prompt}, {"role": "assistant", 'content': message['content']}]
    answer = {'course_id': test['course_id'], 'question_id': test['question_id'], 'chat_id': message['chat_id'], "interaction" : interaction, 
              'confidence' : None}
    all_iteractions.append(answer)
````

````python
with open('m1_submission_358599.json', 'w') as f:
    json.dump(all_iteractions, f)
````
